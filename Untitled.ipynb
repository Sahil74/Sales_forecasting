# Feature Engineering

### Assume our client give some additional information on the Item_identifier column that FD means Food items, DR means Drink items and NC means Non consumable

```python
#see here how we use .str to work each row for slicing vv
x_train_c.Item_Identifier.str[:2].value_counts()
```

```python
# Now we have to chamge ITEM_TYPE as Food Item, Non consumables, Drink by first converting them into FD,NC,DR and then apply map function
# why ITEM_TYPE? = Current values are not effective but if we convert them into these categories then the column would be more important
def convert(data_f):
    data_f['Item_Type'] = data_f['Item_Identifier'].str[:2]
    data_f['Item_Type'] = data_f['Item_Type'].map({'FD' : "Food", 'DR' : "Drink", 'NC':'Non_consumable'})
```

```python
convert(x_train_c)
x_train_c.head()
```

## Missing data

### Item_weight has some missing values - We have item_identifier and there is possibility that same item's weight is defined at one place and not at other so we can use that and fill up null value
### Level - if data is still missing and we can not map = Median

```python
# We can see here
x_train_c[["Item_Identifier","Item_Weight"]].sort_values(by = ["Item_Identifier"])
```

```python
# we are using pivot table to get all uniqe item_id with weitgths
Item_I_W_Pivot = x_train_c.pivot_table(values= "Item_Weight", index = "Item_Identifier").reset_index()
Item_I_W_map = dict(zip(Item_I_W_Pivot["Item_Identifier"],Item_I_W_Pivot["Item_Weight"]))
#Item_I_W_map
```

```python
Item_T_W_Pivot = x_train_c.pivot_table(values= "Item_Weight", index = "Item_Type",aggfunc= 'median').reset_index()
Item_T_W_map = dict(zip(Item_T_W_Pivot["Item_Type"],Item_T_W_Pivot["Item_Weight"]))
```

```python
def Impute_weight(data_f):
    #main
    data_f.loc[:,"Item_Weight"] = data_f.loc[:,"Item_Weight"].fillna(data_f.loc[:,"Item_Identifier"].map(Item_I_W_map))
    # if data is still missing
    data_f.loc[:,"Item_Weight"] = data_f.loc[:,"Item_Weight"].fillna(data_f.loc[:,"Item_Type"].map(Item_T_W_map))
```

```python
Impute_weight(x_train_c)
```

```python
 #Lets do the almost same here
from scipy.stats import mode
#we are using this function becasue in mormal mode, the mode for grocery = "nan" because of missing values
def custom_mode(series):
    series = series.dropna()
    return series.mode()[0] if not series.empty else np.nan
Outlet_T_S_pivot = x_train_c.pivot_table(values = "Outlet_Size", index = "Outlet_Type", aggfunc =custom_mode).reset_index()
Outlet_T_S_map = dict(zip(Outlet_T_S_pivot['Outlet_Type'],Outlet_T_S_pivot['Outlet_Size']))
```

```python
def Impute_OS(data_f):
    data_f.loc[:,"Outlet_Size"] = data_f.loc[:,"Outlet_Size"].fillna(data_f.loc[:,"Outlet_Type"].map(Outlet_T_S_map))
```

```python
def train_model(model, x_train,y_train, cv= 5):
    cv_results = cross_validate(model, x_train, y_train, cv =cv, scoring = ('r2','neg_root_mean_squared_error'),)
    print("model:",model)
    r2_scores = cv_results['test_r2']
    print('R2 cv',r2_scores)
    print("r2 mean/std",np.mean(r2_scores), '/',np.std(r2_scores))
    
    rmse_scores = cv_results['test_neg_root_mean_squared_error']
    rmse_scores =[-1* score for score in rmse_scores]
    print('rmse_scores cv',rmse_scores)
    print("rmse_scores mean/std",np.mean(rmse_scores), '/',np.std(rmse_scores))
```

```python
rf = RandomForestRegressor(random_state= 121)
train_model(rf,x_train = x_train_final,y_train=y_train)
```

```python
x_train_copy = x_train.copy().drop(columns= 'Item_Identifier')
cat_cols = x_train_copy.select_dtypes(include=['object']).columns.tolist()
num_cols = cal_cols = x_train_copy.select_dtypes(exclude=[ 'object']).columns.tolist()
cat_cols, num_cols
```

```python
x_train_copy[cat_cols] = x_train_copy[cat_cols].astype('category')
n_categorical_features = len(cat_cols)
n_numerical_features = len (num_cols)
x_train_copy = x_train_copy[cat_cols+num_cols]
x_train_copy. info()
```

```python
from sklearn.preprocessing import OrdinalEncoder 
from sklearn.pipeline import make_pipeline 
from sklearn.compose import make_column_transformer 
from sklearn. compose import make_column_selector

categorical_mask = [True] * n_categorical_features + [False] * n_numerical_features

ordinal_encoder = make_column_transformer (
        (
        OrdinalEncoder (handle_unknown="use_encoded_value", unknown_value=np.nan), 
        make_column_selector(dtype_include="category"),
        ),
        remainder="passthrough",
)

hgb = make_pipeline(
        ordinal_encoder, 
        HistGradientBoostingRegressor(
            random_state=42, categorical_features=categorical_mask
        ),
)
train_model(model=hgb, x_train=x_train_copy, y_train=y_train)
```

```python
lgbr = LGBMRegressor (random_state=121)
train_model(model=lgbr, x_train=x_train_copy, y_train=y_train)
```
